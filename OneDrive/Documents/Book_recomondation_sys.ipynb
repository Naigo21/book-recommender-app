##**Popularity based recomondation system**





```python
ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.shape
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_books=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_books=popular_books[popular_books['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_books=popular_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

## **Collebrative Filter Based Recomondation**

```python
active_readers=ratings.groupby('User-ID')['Book-Rating'].count().sort_values(ascending=False).reset_index()
active_readers.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
active_readers=active_readers[active_readers.num_ratings>200]
```

```python
active_readers
```

```python
U=ratings_with_name.groupby('User-ID')['Book-Rating'].count()>200
active_readers=U[U].index
```

```python
len(active_readers)
```

```python
filter_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(active_readers)]
```

```python
filter_ratings.shape
```

```python
B=filter_ratings.groupby('Book-Title')['Book-Rating'].count()>=50
famous_books=B[B].index
```

```python
famous_books.shape
```

```python
final_ratings=filter_ratings[filter_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.duplicated().sum()
final_ratings.shape
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.index[0:50]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:7]
    similar_book=[]
    for i in similar_items:
        similar_book.append(pt.index[i[0]])
    return similar_book

```

```python
recommend('The Fellowship of the Ring (The Lord of the Rings, Part 1)')
```
